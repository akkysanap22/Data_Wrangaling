```python
Sales & Marketing Team will be merged
Rename dept column for sales & marketing people to S&M
```

```python
def f(d):
    if (d == 'sales') | (d == 'marketing'):
        return "S&M"
    else:
        return d
    
hr_data['new_dept'] = hr_data.Department.map(f)
```

### Assignments on HR data

```python
1. Convert average_montly_hours to days and hours
```

```python
def convert_hours_to_days(h):
    day = h.average_montly_hours//24
    hour = h.average_montly_hours%24
    return str(day) + ' days ' + str(hour) + ' hours'

hr_data['avg_work_time_per_month'] = hr_data.apply(convert_hours_to_days, axis = 1)
```

```python
Q. Find the rows with maximum number of uppercase words (i.e. first character is uppercase)
```

```python
def num_of_uppercase_word(list_words):
    count = 0
    for item in list_words:
        if 65<=ord(item[0])<=91:
            count += 1
    return count

horror_data['num_of_uppercase_words'] = horror_data.list_words.map(num_of_uppercase_word)
```

```python
def f(r):
    return len(list(filter(lambda w:w.istitle(),r)))
```

```python
horror_data['title_words'] = horror_data.list_words.map(f)
```

```python
horror_data.title_words.max()
```

## Understanding pivot table

```python
gap_data = pd.read_csv("https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/gapminder-FiveYearData.csv")
```

```python
gap_data.head()
```

```python
life_exp_country = gap_data.pivot_table(index='country',columns='year',values='lifeExp')
```

```python
life_exp_country.head()
```

```python
Q. Find out most improved country in terms of life expectency
```

```python
life_exp_country['improvement'] = life_exp_country[2007] - life_exp_country[1952]
```

```python
life_exp_country.head()
```

```python
life_exp_country[life_exp_country.improvement == life_exp_country['improvement'].max()]
```

```python
Q. Calculate life expectency of the continent as mean of all the countries belonging to it.
```

```python
ggap_data = gap_data.groupby(['continent','year'])
```

```python
cont_life_ext = ggap_data.lifeExp.mean()
```

```python
cont_life_ext = pd.DataFrame(cont_life_ext)
```

```python
cont_life_ext.head()
```

```python
Unstacking to convert inner index to column
```

```python
cont_life_ext = cont_life_ext.unstack()
```

```python
cont_life_ext
```

```python
import numpy as np
gap_data.pivot_table(index='continent',columns='year',values='lifeExp',aggfunc=np.sum)
# here by default aggfunc takes mean
```
